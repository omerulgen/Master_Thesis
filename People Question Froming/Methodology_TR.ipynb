```python
import pandas as pd
import random

def generate_choices(correct_answer, options_list, num_choices=3):
    """Generate random choices for a given correct answer."""
    if correct_answer not in options_list:
        options_list.append(correct_answer)
    choices = random.sample([option for option in options_list if option != correct_answer], num_choices)
    choices.append(correct_answer)
    random.shuffle(choices)  # Randomize order of choices
    return choices

def create_quiz_questions(data, name_column, occupation_column, nationality_column, birth_year_column, question_format_nationality, question_format_occupation, question_format_birth_year):
    """
    Create quiz questions for nationality, profession, and birth year for each entry in the dataset.
    
    :param data: DataFrame containing the dataset.
    :param name_column: Column name for the person's name.
    :param occupation_column: Column name for the person's occupation.
    :param nationality_column: Column name for the person's nationality.
    :param birth_year_column: Column name for the person's birth year.
    :param question_format_nationality: Question format string for nationality.
    :param question_format_occupation: Question format string for occupation.
    :param question_format_birth_year: Question format string for birth year.
    :return: Lists of dictionaries containing the questions and options.
    """
    unique_nationalities = data[nationality_column].dropna().unique()
    unique_occupations = data[occupation_column].dropna().unique()
    unique_birth_years = data[birth_year_column].dropna().unique().tolist()

    nationality_questions = []
    profession_questions = []
    birth_year_questions = []

    for index, row in data.iterrows():
        name = row[name_column]
        correct_nationality = row[nationality_column]
        correct_occupation = row[occupation_column]
        correct_birth_year = row[birth_year_column]

        if pd.isna(name) or pd.isna(correct_nationality) or pd.isna(correct_occupation) or pd.isna(correct_birth_year):
            continue

        nationality_choices = generate_choices(correct_nationality, unique_nationalities)
        occupation_choices = generate_choices(correct_occupation, unique_occupations)
        birth_year_choices = generate_choices(correct_birth_year, unique_birth_years, num_choices=3)

        nationality_question = question_format_nationality.format(name=name)
        occupation_question = question_format_occupation.format(name=name)
        birth_year_question = question_format_birth_year.format(name=name)

        # Get correct letter for the correct answer
        correct_letter_nationality = chr(65 + nationality_choices.index(correct_nationality))
        correct_letter_occupation = chr(65 + occupation_choices.index(correct_occupation))
        correct_letter_birth_year = chr(65 + birth_year_choices.index(correct_birth_year))

        nationality_questions.append({'question': nationality_question, 'options': {f'{chr(65+i)}': choice for i, choice in enumerate(nationality_choices)}, 'correct_answer': correct_nationality, 'correct_letter': correct_letter_nationality})
        profession_questions.append({'question': occupation_question, 'options': {f'{chr(65+i)}': choice for i, choice in enumerate(occupation_choices)}, 'correct_answer': correct_occupation, 'correct_letter': correct_letter_occupation})
        birth_year_questions.append({'question': birth_year_question, 'options': {f'{chr(65+i)}': choice for i, choice in enumerate(birth_year_choices)}, 'correct_answer': correct_birth_year, 'correct_letter': correct_letter_birth_year})

    return nationality_questions, profession_questions, birth_year_questions

# Define column names and question formats for each language including English
languages = {
    'EN': ('itemNameEN', 'singleOccupationLabelEN', 'singleNationalityLabelEN', 'birthYear', "What is the nationality of {name}?", "What is the profession of {name}?", "What year was {name} born?"),
    'FR': ('itemNameFR', 'singleOccupationLabelFR', 'singleNationalityLabelFR', 'birthYear', "Quelle est la nationalité de {name}?", "Quelle est la profession de {name}?", "Quelle est l'année de naissance de {name}?"),
    'DE': ('itemNameDE', 'singleOccupationLabelDE', 'singleNationalityLabelDE', 'birthYear', "Was ist die Nationalität von {name}?", "Was ist der Beruf von {name}?", "In welchem Jahr wurde {name} geboren?"),
    'TR': ('itemNameTR', 'singleOccupationLabelTR', 'singleNationalityLabelTR', 'birthYear', "{name} hangi ülkenin vatandaşı?", "{name} mesleği nedir?", "{name} hangi yıl doğdu?"),
    'JA': ('itemNameJA', 'singleOccupationLabelJA', 'singleNationalityLabelJA', 'birthYear', "{name} の国籍は何ですか？", "{name} の職業は何ですか？", "{name} は何年に生まれましたか？")
}

# Generate questions for each language
#for code, (name_col, occ_col, nat_col, birth_col, nat_q, occ_q, birth_q) in languages.items():
#    nat_questions, occ_questions, birth_questions = create_quiz_questions(data, name_col, occ_col, nat_col, birth_col, nat_q, occ_q, birth_q)
#    print(f"{code} - Nationality Questions: {nat_questions[:1]}")
#    print(f"{code} - Profession Questions: {occ_questions[:1]}")
#    print(f"{code} - Birth Year Questions: {birth_questions[:1]}")

# Define a function to nicely print the question and options
def print_question_details(question_details):
    print(f"Question: {question_details['question']}")
    for option_key, option_value in question_details['options'].items():
        print(f"  {option_key}: {option_value}")
    print(f"Correct Answer: {question_details['correct_answer']} (Option {question_details['correct_letter']})")
    print()

# Generate questions for each language and print them in a structured way
for code, (name_col, occ_col, nat_col, birth_col, nat_q, occ_q, birth_q) in languages.items():
    nat_questions, occ_questions, birth_questions = create_quiz_questions(data, name_col, occ_col, nat_col, birth_col, nat_q, occ_q, birth_q)
    
    print(f"===== {code} - Nationality Questions =====")
    if nat_questions:
        print_question_details(nat_questions[0])
    
    print(f"===== {code} - Profession Questions =====")
    if occ_questions:
        print_question_details(occ_questions[0])

    print(f"===== {code} - Birth Year Questions =====")
    if birth_questions:
        print_question_details(birth_questions[0])
```

```python
language_settings = {
    'EN': {
        'name_column': 'itemNameEN',
        'occupation_column': 'singleOccupationLabelEN',
        'nationality_column': 'singleNationalityLabelEN',
        'birth_year_column': 'birthYear',
        'unique_nationalities': data['singleNationalityLabelEN'].dropna().unique(),
        'unique_occupations': data['singleOccupationLabelEN'].dropna().unique(),
        'unique_birth_years': data['birthYear'].dropna().unique()
    },
    'FR': {
        'name_column': 'itemNameFR',
        'occupation_column': 'singleOccupationLabelFR',
        'nationality_column': 'singleNationalityLabelFR',
        'birth_year_column': 'birthYear',
        'unique_nationalities': data['singleNationalityLabelFR'].dropna().unique(),
        'unique_occupations': data['singleOccupationLabelFR'].dropna().unique(),
        'unique_birth_years': data['birthYear'].dropna().unique()
    },
    'DE': {
        'name_column': 'itemNameDE',
        'occupation_column': 'singleOccupationLabelDE',
        'nationality_column': 'singleNationalityLabelDE',
        'birth_year_column': 'birthYear',
        'unique_nationalities': data['singleNationalityLabelDE'].dropna().unique(),
        'unique_occupations': data['singleOccupationLabelDE'].dropna().unique(),
        'unique_birth_years': data['birthYear'].dropna().unique()
    },
    'TR': {
        'name_column': 'itemNameTR',
        'occupation_column': 'singleOccupationLabelTR',
        'nationality_column': 'singleNationalityLabelTR',
        'birth_year_column': 'birthYear',
        'unique_nationalities': data['singleNationalityLabelTR'].dropna().unique(),
        'unique_occupations': data['singleOccupationLabelTR'].dropna().unique(),
        'unique_birth_years': data['birthYear'].dropna().unique()
    },
    'JA': {
        'name_column': 'itemNameJA',
        'occupation_column': 'singleOccupationLabelJA',
        'nationality_column': 'singleNationalityLabelJA',
        'birth_year_column': 'birthYear',
        'unique_nationalities': data['singleNationalityLabelJA'].dropna().unique(),
        'unique_occupations': data['singleOccupationLabelJA'].dropna().unique(),
        'unique_birth_years': data['birthYear'].dropna().unique()
    }
}
def generate_language_specific_questions(name, nationality, occupation, birth_year, unique_nationalities, unique_occupations, unique_birth_years, language):
    result = []

    # Generate nationality question
    nationality_true = random.choice([True, False])
    if nationality_true:
        question_nationality = {
            'EN': f"{name} is from {nationality}.",
            'FR': f"{name} est originaire de {nationality}.",
            'DE': f"{name} kommt aus {nationality}.",
            'TR': f"{name} {nationality}'lı.",
            'JA': f"{name} は {nationality} 出身です。"
        }[language]
        correct_answer_nationality = True
        real_answer_nationality = nationality
    else:
        fake_nationality = random.choice([nat for nat in unique_nationalities if nat != nationality])
        question_nationality = {
            'EN': f"{name} is from {fake_nationality}.",
            'FR': f"{name} est originaire de {fake_nationality}.",
            'DE': f"{name} kommt aus {fake_nationality}.",
            'TR': f"{name} {fake_nationality}'lı.",
            'JA': f"{name} は {fake_nationality} 出身です。"
        }[language]
        correct_answer_nationality = False
        real_answer_nationality = nationality

    # Generate occupation question
    occupation_true = random.choice([True, False])
    if occupation_true:
        question_occupation = {
            'EN': f"{name} is a(n) {occupation}.",
            'FR': f"{name} est un(e) {occupation}.",
            'DE': f"{name} ist ein(e) {occupation}.",
            'TR': f"{name} bir {occupation}.",
            'JA': f"{name} は {occupation} です。"
        }[language]
        correct_answer_occupation = True
        real_answer_occupation = occupation
    else:
        fake_occupation = random.choice([occ for occ in unique_occupations if occ != occupation])
        question_occupation = {
            'EN': f"{name} is a(n) {fake_occupation}.",
            'FR': f"{name} est un(e) {fake_occupation}.",
            'DE': f"{name} ist ein(e) {fake_occupation}.",
            'TR': f"{name} bir {fake_occupation}.",
            'JA': f"{name} は {fake_occupation} です。"
        }[language]
        correct_answer_occupation = False
        real_answer_occupation = occupation

    # Generate birth year question
    birth_year_true = random.choice([True, False])
    if birth_year_true:
        question_birth_year = {
            'EN': f"{name} was born in {birth_year}.",
            'FR': f"{name} est né en {birth_year}.",
            'DE': f"{name} wurde {birth_year} geboren.",
            'TR': f"{name} {birth_year} yılında doğdu.",
            'JA': f"{name} は {birth_year} 年に生まれました。"
        }[language]
        correct_answer_birth_year = True
        real_answer_birth_year = birth_year
    else:
        fake_birth_year = random.choice([year for year in unique_birth_years if year != birth_year])
        question_birth_year = {
            'EN': f"{name} was born in {fake_birth_year}.",
            'FR': f"{name} est né en {fake_birth_year}.",
            'DE': f"{name} wurde {fake_birth_year} geboren.",
            'TR': f"{name} {fake_birth_year} yılında doğdu.",
            'JA': f"{name} は {fake_birth_year} 年に生まれました。"
        }[language]
        correct_answer_birth_year = False
        real_answer_birth_year = birth_year

    result.extend([
        {'question': question_nationality, 'correct_answer': correct_answer_nationality, 'real_answer': real_answer_nationality},
        {'question': question_occupation, 'correct_answer': correct_answer_occupation, 'real_answer': real_answer_occupation},
        {'question': question_birth_year, 'correct_answer': correct_answer_birth_year, 'real_answer': real_answer_birth_year}
    ])

    return result

# Define the function to generate true/false questions
def generate_true_false_questions(data, settings):
    questions = {lang: [] for lang in settings}
    for index, row in data.iterrows():
        for lang, lang_settings in settings.items():
            name = row[lang_settings['name_column']]
            actual_nationality = row[lang_settings['nationality_column']]
            actual_occupation = row[lang_settings['occupation_column']]
            actual_birth_year = row[lang_settings['birth_year_column']]
            
            if pd.isna(name) or pd.isna(actual_nationality) or pd.isna(actual_occupation) or pd.isna(actual_birth_year):
                continue  # Skip incomplete data entries

            questions[lang].extend(generate_language_specific_questions(
                name, actual_nationality, actual_occupation, actual_birth_year,
                lang_settings['unique_nationalities'],
                lang_settings['unique_occupations'],
                lang_settings['unique_birth_years'],
                lang
            ))

    return questions

# Now call the function using the defined settings
true_false_questions = generate_true_false_questions(data, language_settings)
#for language, qs in true_false_questions.items():
#    print(f"{language} - Sample Questions:")
#    for question in qs[6:9]:  # Print the first three questions for each language
#        print(question)
        
# Define a function to nicely print the true/false question details
def print_question_details(question):
    print(f"Question: {question['question']}")
    print(f"True/False: {'True' if question['correct_answer'] else 'False'}")
    print(f"Actual Answer: {question['real_answer']}")
    print()

# Define the function to generate true/false questions and print them
def generate_and_print_true_false_questions(data, settings):
    questions = {lang: [] for lang in settings}
    for index, row in data.iterrows():
        for lang, lang_settings in settings.items():
            name = row[lang_settings['name_column']]
            actual_nationality = row[lang_settings['nationality_column']]
            actual_occupation = row[lang_settings['occupation_column']]
            actual_birth_year = row[lang_settings['birth_year_column']]
            
            if pd.isna(name) or pd.isna(actual_nationality) or pd.isna(actual_occupation) or pd.isna(actual_birth_year):
                continue  # Skip incomplete data entries

            questions[lang].extend(generate_language_specific_questions(
                name, actual_nationality, actual_occupation, actual_birth_year,
                lang_settings['unique_nationalities'],
                lang_settings['unique_occupations'],
                lang_settings['unique_birth_years'],
                lang
            ))
    
    # Now print the formatted questions for each language
    for language, qs in questions.items():
        print(f"===== {language} - Sample Questions: =====")
        for question in qs[:3]:  # Adjust the range for the number of questions you want to print
            print_question_details(question)

# Call the function using the defined settings
generate_and_print_true_false_questions(data, language_settings)
```

```python
import pandas as pd

def generate_open_ended_questions(name, nationality, occupation, birth_year, language):
    # Open-ended questions in different languages
    question_templates = {
        'EN': {
            'nationality': f"What is the nationality of {name}?",
            'occupation': f"What is the occupation of {name}?",
            'birth_year': f"In what year was {name} born?"
        },
        'FR': {
            'nationality': f"Quelle est la nationalité de {name}?",
            'occupation': f"Quelle est la profession de {name}?",
            'birth_year': f"En quelle année {name} est-il né?"
        },
        'DE': {
            'nationality': f"Was ist die Nationalität von {name}?",
            'occupation': f"Was ist der Beruf von {name}?",
            'birth_year': f"In welchem Jahr wurde {name} geboren?"
        },
        'TR': {
            'nationality': f"{name} hangi ülkenin vatandaşıdır?",
            'occupation': f"{name} mesleği nedir?",
            'birth_year': f"{name} hangi yıl doğmuştur?"
        },
        'JA': {
            'nationality': f"{name}の国籍は何ですか？",
            'occupation': f"{name}の職業は何ですか？",
            'birth_year': f"{name}は何年に生まれましたか？"
        }
    }

    # Generate open-ended questions
    question_nationality = question_templates[language]['nationality']
    question_occupation = question_templates[language]['occupation']
    question_birth_year = question_templates[language]['birth_year']

    # Output structure includes only the questions and correct answers
    result = [
        {'question': question_nationality, 'correct_answer': nationality},
        {'question': question_occupation, 'correct_answer': occupation},
        {'question': question_birth_year, 'correct_answer': birth_year}
    ]

    return result

# Define the function to generate all open-ended questions for the dataset
def generate_all_open_ended_questions(data, settings):
    all_questions = {lang: [] for lang in settings}
    for index, row in data.iterrows():
        for lang, lang_settings in settings.items():
            name = row[lang_settings['name_column']]
            nationality = row[lang_settings['nationality_column']]
            occupation = row[lang_settings['occupation_column']]
            birth_year = row[lang_settings['birth_year_column']]

            if pd.isna(name) or pd.isna(nationality) or pd.isna(occupation) or pd.isna(birth_year):
                continue  # Skip incomplete data entries

            all_questions[lang].extend(generate_open_ended_questions(
                name, nationality, occupation, birth_year, lang
            ))

    return all_questions

# Define a function to nicely print the open-ended question details
def print_open_ended_question_details(question):
    print(f"Question: {question['question']}")
    print(f"Correct Answer: {question['correct_answer']}")
    print()

# Define the function to generate, print, and collect all open-ended questions for the dataset
def generate_and_print_all_open_ended_questions(data, settings):
    all_questions = {lang: [] for lang in settings}
    questions_list = []
    for index, row in data.iterrows():
        for lang, lang_settings in settings.items():
            name = row[lang_settings['name_column']]
            nationality = row[lang_settings['nationality_column']]
            occupation = row[lang_settings['occupation_column']]
            birth_year = row[lang_settings['birth_year_column']]

            if pd.isna(name) or pd.isna(nationality) or pd.isna(occupation) or pd.isna(birth_year):
                continue  # Skip incomplete data entries

            questions = generate_open_ended_questions(name, nationality, occupation, birth_year, lang)
            all_questions[lang].extend(questions)
            for question in questions:
                questions_list.append({
                    'language': lang,
                    'question': question['question'],
                    'correct_answer': question['correct_answer']
                })

    # Now print the formatted questions for each language
    for language, qs in all_questions.items():
        print(f"===== {language} - Sample Open-Ended Questions: =====")
        for question in qs[:3]:  # Adjust the range for the number of questions you want to print
            print_open_ended_question_details(question)
        print()
    
    # Create a DataFrame from the collected questions
    all_questions_oe = pd.DataFrame(questions_list)
    return all_questions_oe


# Call the function using the defined settings
all_questions_oe = generate_and_print_all_open_ended_questions(data, language_settings)

# Display the resulting DataFrame
#print(all_questions_oe)
```
