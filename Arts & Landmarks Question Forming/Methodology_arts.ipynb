```python
import pandas as pd
import random

def generate_choices(correct_answer, options_list, num_choices=3, is_year=False):
    """Generate random choices for a given correct answer."""
    if correct_answer not in options_list:
        options_list.append(correct_answer)
    if is_year:
        options_list = [str(option)[:4] for option in options_list]
    choices = random.sample([option for option in options_list if option != correct_answer], num_choices)
    choices.append(correct_answer)
    random.shuffle(choices)  # Randomize order of choices
    return choices

def extract_year(date_str):
    """Extract year from date string."""
    if pd.isna(date_str):
        return None
    return date_str[:4]

def create_quiz_questions(data, name_columns, creator_columns, year_column, question_format_creator, question_format_year):
    """
    Create quiz questions for creator and year for each entry in the dataset.
    
    :param data: DataFrame containing the dataset.
    :param name_columns: Dictionary of column names for the item's name in different languages.
    :param creator_columns: Dictionary of column names for the item's creator in different languages.
    :param year_column: Column name for the item's year.
    :param question_format_creator: Dictionary of question format strings for creator.
    :param question_format_year: Dictionary of question format strings for year.
    :return: Dictionary of questions for each language.
    """
    # Combine all unique creators from different language columns
    unique_creators = pd.concat([data[col] for col in creator_columns.values()]).dropna().unique()
    unique_years = [extract_year(date) for date in data[year_column].dropna().unique()]

    questions = {lang: [] for lang in name_columns.keys()}

    for index, row in data.iterrows():
        name_entries = {lang: row[col] for lang, col in name_columns.items()}
        creator_entries = {lang: row[col] for lang, col in creator_columns.items()}
        correct_year = extract_year(row[year_column])

        if any(pd.isna(list(name_entries.values()))) or any(pd.isna(list(creator_entries.values()))) or pd.isna(correct_year):
            continue

        creator_choices = generate_choices(creator_entries['EN'], unique_creators)
        year_choices = generate_choices(correct_year, unique_years, num_choices=3, is_year=True)

        for lang in name_columns.keys():
            creator_question = question_format_creator[lang].format(name=name_entries[lang])
            year_question = question_format_year[lang].format(name=name_entries[lang])

            correct_letter_creator = chr(65 + creator_choices.index(creator_entries['EN']))
            correct_letter_year = chr(65 + year_choices.index(correct_year))

            questions[lang].append({
                'creator_question': {'question': creator_question, 'options': {f'{chr(65+i)}': choice for i, choice in enumerate(creator_choices)}, 'correct_answer': creator_entries['EN'], 'correct_letter': correct_letter_creator},
                'year_question': {'question': year_question, 'options': {f'{chr(65+i)}': choice for i, choice in enumerate(year_choices)}, 'correct_answer': correct_year, 'correct_letter': correct_letter_year}
            })

    return questions

# Define column names and question formats for each language
name_columns = {
    'EN': 'placeLabelEN',
    'FR': 'placeLabelFR',
    'DE': 'placeLabelDE',
    'TR': 'placeLabelTR',
    'JA': 'placeLabelJA'
}
creator_columns = {
    'EN': 'creatorLabelEN',
    'FR': 'creatorLabelFR',
    'DE': 'creatorLabelDE',
    'TR': 'creatorLabelTR',
    'JA': 'creatorLabelJA'
}
year_column = 'dateCompleted'
question_format_creator = {
    'EN': "Who is the creator of {name}?",
    'FR': "Qui est le créateur de {name}?",
    'DE': "Wer ist der Schöpfer von {name}?",
    'TR': "{name}'in yaratıcısı kim?",
    'JA': "{name} の創作者は誰ですか？"
}
question_format_year = {
    'EN': "In which year was {name} completed?",
    'FR': "En quelle année {name} a-t-il été achevé?",
    'DE': "In welchem Jahr wurde {name} fertiggestellt?",
    'TR': "{name} hangi yılda tamamlandı?",
    'JA': "{name} は何年に完成しましたか？"
}

# Generate questions for each language
questions = create_quiz_questions(filtered_df, name_columns, creator_columns, year_column, question_format_creator, question_format_year)

# Define a function to nicely print the question and options
def print_question_details(question_details):
    print(f"Question: {question_details['question']}")
    for option_key, option_value in question_details['options'].items():
        print(f"  {option_key}: {option_value}")
    print(f"Correct Answer: {question_details['correct_answer']} (Option {question_details['correct_letter']})")
    print()

# Print the generated questions for verification
for lang, qs in questions.items():
    print(f"===== {lang} - Sample Questions =====")
    if qs:
        print("===== Creator Question =====")
        print_question_details(qs[3000]['creator_question'])
        print("===== Year Question =====")
        print_question_details(qs[3000]['year_question'])
    print()
```

## Now we move on to the second format of questions. 

### This format will be in the form of True/False statements where 50% of questions framed will be True.

```python
import pandas as pd
import random

def generate_language_specific_questions(name, creator, year, unique_creators, unique_years, language):
    result = []

    # Generate creator question
    creator_true = random.choice([True, False])
    if creator_true:
        question_creator = {
            'EN': f"Was {name} created by {creator}?",
            'FR': f"{name} a-t-il été créé par {creator}?",
            'DE': f"Wurde {name} von {creator} erstellt?",
            'TR': f"{name} {creator} tarafından mı yaratıldı?",
            'JA': f"{name} は {creator} によって作られましたか？"
        }[language]
        correct_answer_creator = True
        real_answer_creator = creator
    else:
        fake_creator = random.choice([c for c in unique_creators if c != creator])
        question_creator = {
            'EN': f"Was {name} created by {fake_creator}?",
            'FR': f"{name} a-t-il été créé par {fake_creator}?",
            'DE': f"Wurde {name} von {fake_creator} erstellt?",
            'TR': f"{name} {fake_creator} tarafından mı yaratıldı?",
            'JA': f"{name} は {fake_creator} によって作られましたか？"
        }[language]
        correct_answer_creator = False
        real_answer_creator = creator

    # Generate year question
    year_true = random.choice([True, False])
    if year_true:
        question_year = {
            'EN': f"Was {name} completed in {year}?",
            'FR': f"{name} a-t-il été achevé en {year}?",
            'DE': f"Wurde {name} im Jahr {year} fertiggestellt?",
            'TR': f"{name} {year} yılında mı tamamlandı?",
            'JA': f"{name} は {year} 年に完成しましたか？"
        }[language]
        correct_answer_year = True
        real_answer_year = year
    else:
        fake_year = random.choice([y for y in unique_years if y != year])
        question_year = {
            'EN': f"Was {name} completed in {fake_year}?",
            'FR': f"{name} a-t-il été achevé en {fake_year}?",
            'DE': f"Wurde {name} im Jahr {fake_year} fertiggestellt?",
            'TR': f"{name} {fake_year} yılında mı tamamlandı?",
            'JA': f"{name} は {fake_year} 年に完成しましたか？"
        }[language]
        correct_answer_year = False
        real_answer_year = year

    result.extend([
        {'question': question_creator, 'correct_answer': correct_answer_creator, 'real_answer': real_answer_creator},
        {'question': question_year, 'correct_answer': correct_answer_year, 'real_answer': real_answer_year}
    ])

    return result

# Define the function to generate true/false questions
def generate_true_false_questions(data, settings):
    questions = {lang: [] for lang in settings}
    for index, row in data.iterrows():
        for lang, lang_settings in settings.items():
            name = row[lang_settings['name_column']]
            actual_creator = row[lang_settings['creator_column']]
            actual_year = row[lang_settings['year_column']]
            
            if pd.isna(name) or pd.isna(actual_creator) or pd.isna(actual_year):
                continue  # Skip incomplete data entries

            questions[lang].extend(generate_language_specific_questions(
                name, actual_creator, actual_year,
                lang_settings['unique_creators'],
                lang_settings['unique_years'],
                lang
            ))

    return questions

# Define column names and question formats for each language including English
languages_settings = {
    'EN': {
        'name_column': 'placeLabelEN',
        'creator_column': 'creatorLabelEN',
        'year_column': 'dateCompleted',
        'unique_creators': filtered_df['creatorLabelEN'].dropna().unique(),
        'unique_years': filtered_df['dateCompleted'].dropna().apply(extract_year).unique()
    },
    'FR': {
        'name_column': 'placeLabelFR',
        'creator_column': 'creatorLabelFR',
        'year_column': 'dateCompleted',
        'unique_creators': filtered_df['creatorLabelFR'].dropna().unique(),
        'unique_years': filtered_df['dateCompleted'].dropna().apply(extract_year).unique()
    },
    'DE': {
        'name_column': 'placeLabelDE',
        'creator_column': 'creatorLabelDE',
        'year_column': 'dateCompleted',
        'unique_creators': filtered_df['creatorLabelDE'].dropna().unique(),
        'unique_years': filtered_df['dateCompleted'].dropna().apply(extract_year).unique()
    },
    'TR': {
        'name_column': 'placeLabelTR',
        'creator_column': 'creatorLabelTR',
        'year_column': 'dateCompleted',
        'unique_creators': filtered_df['creatorLabelTR'].dropna().unique(),
        'unique_years': filtered_df['dateCompleted'].dropna().apply(extract_year).unique()
    },
    'JA': {
        'name_column': 'placeLabelJA',
        'creator_column': 'creatorLabelJA',
        'year_column': 'dateCompleted',
        'unique_creators': filtered_df['creatorLabelJA'].dropna().unique(),
        'unique_years': filtered_df['dateCompleted'].dropna().apply(extract_year).unique()
    }
}

# Now call the function using the defined settings
true_false_questions = generate_true_false_questions(filtered_df, languages_settings)

# Define a function to nicely print the true/false question details
def print_question_details(question):
    print(f"Question: {question['question']}")
    print(f"True/False: {'True' if question['correct_answer'] else 'False'}")
    print(f"Actual Answer: {question['real_answer']}")
    print()

# Define the function to generate true/false questions and print them
def generate_and_print_true_false_questions(data, settings):
    questions = {lang: [] for lang in settings}
    for index, row in data.iterrows():
        for lang, lang_settings in settings.items():
            name = row[lang_settings['name_column']]
            actual_creator = row[lang_settings['creator_column']]
            actual_year = row[lang_settings['year_column']]

            if pd.isna(name) or pd.isna(actual_creator) or pd.isna(actual_year):
                continue  # Skip incomplete data entries

            questions[lang].extend(generate_language_specific_questions(
                name, actual_creator, extract_year(actual_year),
                lang_settings['unique_creators'],
                lang_settings['unique_years'],
                lang
            ))

    # Now print the formatted questions for each language
    for language, qs in questions.items():
        print(f"===== {language} - Sample Questions: =====")
        for question in qs[:3]:  # Adjust the range for the number of questions you want to print
            print_question_details(question)

# Call the function using the defined settings
generate_and_print_true_false_questions(filtered_df, languages_settings)
```

```python
import csv

def generate_true_false_questions_to_csv(data, settings, output_file):
    questions = {lang: [] for lang in settings}
    
    # Collect all questions
    for index, row in data.iterrows():
        for lang, lang_settings in settings.items():
            name = row[lang_settings['name_column']]
            actual_creator = row[lang_settings['creator_column']]
            actual_year = row[lang_settings['year_column']]
            
            if pd.isna(name) or (pd.isna(actual_creator) and pd.isna(actual_year)):
                continue  # Skip incomplete data entries

            questions[lang].extend(generate_language_specific_questions(
                name, actual_creator, extract_year(actual_year),
                lang_settings['unique_creators'],
                lang_settings['unique_years'],
                lang
            ))
    
    # Write the questions to a CSV file
    with open(output_file, mode='w', newline='', encoding='utf-8') as file:
        writer = csv.writer(file)
        writer.writerow(['question', 'True/False', 'ActualAnswer', 'Language'])
        
        for language, qs in questions.items():
            for question in qs:
                writer.writerow([question['question'], 'True' if question['correct_answer'] else 'False', question['real_answer'], language])

# Define column names and question formats for each language
languages_settings = {
    'EN': {
        'name_column': 'placeLabelEN',
        'creator_column': 'creatorLabelEN',
        'year_column': 'dateCompleted',
        'unique_creators': data['creatorLabelEN'].dropna().unique(),
        'unique_years': data['dateCompleted'].dropna().apply(extract_year).unique()
    },
    'FR': {
        'name_column': 'placeLabelFR',
        'creator_column': 'creatorLabelFR',
        'year_column': 'dateCompleted',
        'unique_creators': data['creatorLabelFR'].dropna().unique(),
        'unique_years': data['dateCompleted'].dropna().apply(extract_year).unique()
    },
    'DE': {
        'name_column': 'placeLabelDE',
        'creator_column': 'creatorLabelDE',
        'year_column': 'dateCompleted',
        'unique_creators': data['creatorLabelDE'].dropna().unique(),
        'unique_years': data['dateCompleted'].dropna().apply(extract_year).unique()
    },
    'TR': {
        'name_column': 'placeLabelTR',
        'creator_column': 'creatorLabelTR',
        'year_column': 'dateCompleted',
        'unique_creators': data['creatorLabelTR'].dropna().unique(),
        'unique_years': data['dateCompleted'].dropna().apply(extract_year).unique()
    },
    'JA': {
        'name_column': 'placeLabelJA',
        'creator_column': 'creatorLabelJA',
        'year_column': 'dateCompleted',
        'unique_creators': data['creatorLabelJA'].dropna().unique(),
        'unique_years': data['dateCompleted'].dropna().apply(extract_year).unique()
    }
}

# Define the output file name
output_file = 'true_false_questions_arts.csv'

# Call the function to generate the questions and export them to CSV
generate_true_false_questions_to_csv(filtered_df, languages_settings, output_file)

print(f"Questions exported to {output_file}")
```

```python
import pandas as pd

def generate_open_ended_questions(name, creator, year, language):
    # Open-ended questions in different languages
    question_templates = {
        'EN': {
            'creator': f"Who is the creator of {name}?",
            'year': f"In which year was {name} completed?"
        },
        'FR': {
            'creator': f"Qui est le créateur de {name}?",
            'year': f"En quelle année {name} a-t-il été achevé?"
        },
        'DE': {
            'creator': f"Wer ist der Schöpfer von {name}?",
            'year': f"In welchem Jahr wurde {name} fertiggestellt?"
        },
        'TR': {
            'creator': f"{name}'in yaratıcısı kim?",
            'year': f"{name} hangi yılda tamamlandı?"
        },
        'JA': {
            'creator': f"{name} の創作者は誰ですか？",
            'year': f"{name} は何年に完成しましたか？"
        }
    }

    # Generate open-ended questions
    question_creator = question_templates[language]['creator']
    question_year = question_templates[language]['year']

    # Output structure includes only the questions and correct answers
    result = []
    if pd.notna(creator):
        result.append({'question': question_creator, 'correct_answer': creator})
    if pd.notna(year):
        result.append({'question': question_year, 'correct_answer': year})

    return result

# Define the function to generate all open-ended questions for the dataset
def generate_all_open_ended_questions(data, settings):
    all_questions = {lang: [] for lang in settings}
    questions_list = []
    
    for index, row in data.iterrows():
        for lang, lang_settings in settings.items():
            name = row[lang_settings['name_column']]
            creator = row[lang_settings['creator_column']]
            year = extract_year(row[lang_settings['year_column']])

            if pd.isna(name):
                continue  # Skip incomplete data entries

            questions = generate_open_ended_questions(name, creator, year, lang)
            all_questions[lang].extend(questions)
            for question in questions:
                questions_list.append({
                    'language': lang,
                    'question': question['question'],
                    'correct_answer': question['correct_answer']
                })

    # Create a DataFrame from the collected questions
    all_questions_oe = pd.DataFrame(questions_list)
    return all_questions_oe

# Define a function to nicely print the open-ended question details
def print_open_ended_question_details(question):
    print(f"Question: {question['question']}")
    print(f"Correct Answer: {question['correct_answer']}")
    print()

# Define the function to generate, print, and collect all open-ended questions for the dataset
def generate_and_print_all_open_ended_questions(data, settings):
    all_questions = generate_all_open_ended_questions(data, settings)
    
    # Now print the formatted questions for each language
    for language in settings.keys():
        print(f"===== {language} - Sample Open-Ended Questions: =====")
        sample_questions = all_questions[all_questions['language'] == language].head(3)
        for _, question in sample_questions.iterrows():
            print_open_ended_question_details(question)
        print()
    
    return all_questions

# Define column names and question formats for each language
languages_settings = {
    'EN': {
        'name_column': 'placeLabelEN',
        'creator_column': 'creatorLabelEN',
        'year_column': 'dateCompleted'
    },
    'FR': {
        'name_column': 'placeLabelFR',
        'creator_column': 'creatorLabelFR',
        'year_column': 'dateCompleted'
    },
    'DE': {
        'name_column': 'placeLabelDE',
        'creator_column': 'creatorLabelDE',
        'year_column': 'dateCompleted'
    },
    'TR': {
        'name_column': 'placeLabelTR',
        'creator_column': 'creatorLabelTR',
        'year_column': 'dateCompleted'
    },
    'JA': {
        'name_column': 'placeLabelJA',
        'creator_column': 'creatorLabelJA',
        'year_column': 'dateCompleted'
    }
}

# Call the function using the defined settings and your DataFrame
all_questions_oe = generate_and_print_all_open_ended_questions(filtered_df, languages_settings)

# Display the resulting DataFrame
print(all_questions_oe.head())
```

```python
def generate_choices(correct_answer, options_list, num_choices=3, is_year=False):
    """Generate random choices for a given correct answer."""
    if correct_answer not in options_list:
        options_list.append(correct_answer)
    if is_year:
        options_list = [str(option)[:4] for option in options_list]
    choices = random.sample([option for option in options_list if option != correct_answer], num_choices)
    choices.append(correct_answer)
    random.shuffle(choices)  # Randomize order of choices
    return choices

def extract_year(date_str):
    """Extract year from date string."""
    if pd.isna(date_str):
        return None
    return date_str[:4]

def create_quiz_questions(data, name_columns, creator_columns, year_column, question_format_creator, question_format_year):
    """
    Create quiz questions for creator and year for each entry in the dataset.
    
    :param data: DataFrame containing the dataset.
    :param name_columns: Dictionary of column names for the item's name in different languages.
    :param creator_columns: Dictionary of column names for the item's creator in different languages.
    :param year_column: Column name for the item's year.
    :param question_format_creator: Dictionary of question format strings for creator.
    :param question_format_year: Dictionary of question format strings for year.
    :return: Dictionary of DataFrames containing the questions for each language.
    """
    unique_creators = pd.concat([data[col] for col in creator_columns.values()]).dropna().unique()
    unique_years = [extract_year(date) for date in data[year_column].dropna().unique()]

    creator_questions = {lang: [] for lang in name_columns.keys()}
    year_questions = {lang: [] for lang in name_columns.keys()}

    for index, row in data.iterrows():
        name_entries = {lang: row[col] for lang, col in name_columns.items()}
        creator_entries = {lang: row[col] for lang, col in creator_columns.items()}
        correct_year = extract_year(row[year_column])

        if any(pd.isna(list(name_entries.values()))):
            continue

        if not any(pd.isna(list(creator_entries.values()))):
            creator_choices = generate_choices(creator_entries['EN'], unique_creators)
            for lang in name_columns.keys():
                creator_question = question_format_creator[lang].format(name=name_entries[lang])
                correct_letter_creator = chr(65 + creator_choices.index(creator_entries['EN']))
                creator_questions[lang].append({
                    'question': creator_question,
                    'options': {f'{chr(65+i)}': choice for i, choice in enumerate(creator_choices)},
                    'correct_letter': correct_letter_creator,
                    'language': lang
                })

        if not pd.isna(correct_year):
            year_choices = generate_choices(correct_year, unique_years, num_choices=3, is_year=True)
            for lang in name_columns.keys():
                year_question = question_format_year[lang].format(name=name_entries[lang])
                correct_letter_year = chr(65 + year_choices.index(correct_year))
                year_questions[lang].append({
                    'question': year_question,
                    'options': {f'{chr(65+i)}': choice for i, choice in enumerate(year_choices)},
                    'correct_letter': correct_letter_year,
                    'language': lang
                })

    creator_dfs = {lang: pd.DataFrame(questions) for lang, questions in creator_questions.items()}
    year_dfs = {lang: pd.DataFrame(questions) for lang, questions in year_questions.items()}

    return creator_dfs, year_dfs

# Define column names and question formats for each language
name_columns = {
    'EN': 'placeLabelEN',
    'FR': 'placeLabelFR',
    'DE': 'placeLabelDE',
    'TR': 'placeLabelTR',
    'JA': 'placeLabelJA'
}
creator_columns = {
    'EN': 'creatorLabelEN',
    'FR': 'creatorLabelFR',
    'DE': 'creatorLabelDE',
    'TR': 'creatorLabelTR',
    'JA': 'creatorLabelJA'
}
year_column = 'dateCompleted'
question_format_creator = {
    'EN': "Who is the creator of {name}?",
    'FR': "Qui est le créateur de {name}?",
    'DE': "Wer ist der Schöpfer von {name}?",
    'TR': "{name}'in yaratıcısı kim?",
    'JA': "{name} の創作者は誰ですか？"
}
question_format_year = {
    'EN': "In which year was {name} completed?",
    'FR': "En quelle année {name} a-t-il été achevé?",
    'DE': "In welchem Jahr wurde {name} fertiggestellt?",
    'TR': "{name} hangi yılda tamamlandı?",
    'JA': "{name} は何年に完成しましたか？"
}

# Generate questions for each language
creator_dfs, year_dfs = create_quiz_questions(filtered_df, name_columns, creator_columns, year_column, question_format_creator, question_format_year)
```
